```python
from enum import Enum, auto
class Class(Enum):
    PLUS = auto()
    MINUS = auto()
    STAR = auto()
    FWDSLASH = auto()
    PERCENT = auto()
    
    MOD = auto()  
    DIV = auto()     
    ORD = auto() #ORD - Vraća redni broj tipa.

    OR = auto()
    AND = auto()
    NOT = auto()
    XOR = auto() # 0110

    EQ = auto()
    NEQ = auto() # <>
    LT = auto()
    GT = auto()
    LTE = auto()
    GTE = auto()

    LPAREN = auto() #()
    RPAREN = auto()
    LBRACKET = auto() #[]
    RBRACKET = auto()
    LBRACE = auto() #{}
    RBRACE = auto()

    ASSIGN = auto()
    SEMICOLON = auto()
    COMMA = auto()
    DOT = auto()
    DOTDOT = auto()
    COLON = auto()

    TYPE = auto()
    INT = auto()
    CHAR = auto()
    STRING = auto()
    REAl = auto()
    BOOLEAN = auto()
 
    IF = auto()
    THEN = auto()
    ELSE = auto()
    WHILE = auto()
    REPEAT = auto()
    UNTIL = auto()
    OF = auto()
    
    FOR = auto()
    TO = auto()
    DO = auto()
    
    BEGIN = auto()
    END = auto()
    VAR = auto()
    EXIT = auto()
    PROCEDURE = auto()
    FUNCTION = auto()
    ARRAY = auto()
    DOWNTO = auto()

    BREAK = auto()
    CONTINUE = auto()
    RETURN = auto()
    
    ADDRESS = auto()
    
    ID = auto()
    EOF = auto()
```

```python
class Token:
    def __init__(self, class_, lexeme):
        self.class_ = class_
        self.lexeme = lexeme

    def __str__(self):
        return "<{} {}>".format(self.class_, self.lexeme)
```

```python
class Lexer:
    def __init__(self, text):
        self.text = text
        self.len = len(text)
        self.pos = -1

#Metoda read_space ne konstruiše token, ali pomera pokazivač na prvi sledeći karakter koji nije razmak.
    def read_space(self):
        while self.pos + 1 < self.len and self.text[self.pos + 1].isspace():
            self.next_char()

#Metoda read_int konstruiše token literala celog broja pod uslovom da je trenutni karakter cifra.
    def read_int(self):
        lexeme = self.text[self.pos]
        while self.pos + 1 < self.len and self.text[self.pos + 1].isdigit():
            lexeme += self.next_char()
        return lexeme

    #def read_real(self):
    #    lexeme = self.text[self.pos]
    #    while self.pos + 1 < self.len and self.text[self.pos + 1].isdigit():
    #        lexeme += self.next_char()
    #    return real(lexeme)

#Metoda read_char konstruiše token literala karaktera pod uslovom da je trenutni karakter apostrof.
    def read_char(self):
        self.pos += 1
        lexeme = self.text[self.pos]
        self.pos += 1
        return lexeme

#Metoda read_string konstruiše token string literala pod uslovom da je trenutni karakter znak navodnika.
    def read_string(self):
        lexeme = ''
        while self.pos + 1 < self.len and self.text[self.pos + 1] != '\'':
            lexeme += self.next_char()
        self.pos += 1
        return lexeme

#Metoda read_keyword konstruiše token ključne reči pod uslovom da je trenutni karakter slovo.
    def read_keyword(self):
        lexeme = self.text[self.pos]
        while self.pos + 1 < self.len and self.text[self.pos + 1].isalnum() or self.text[self.pos + 1] == '_':
            lexeme += self.next_char()
        #while self.text[self.pos + 1] == '_':
        #    lexeme += self.next_char()
        if lexeme == 'if':
            return Token(Class.IF, lexeme)
        if lexeme == 'then':
            return Token(Class.THEN, lexeme)
        elif lexeme == 'else':
            return Token(Class.ELSE, lexeme)
        elif lexeme == 'while':
            return Token(Class.WHILE, lexeme)
        elif lexeme == 'repeat':
            return Token(Class.REPEAT, lexeme)
        elif lexeme == 'until':
            return Token(Class.UNTIL, lexeme)
        elif lexeme == 'of':
            return Token(Class.OF, lexeme)
        elif lexeme == 'for':
            return Token(Class.FOR, lexeme)
        elif lexeme == 'to':
            return Token(Class.TO, lexeme)
        elif lexeme == 'do':
            return Token(Class.DO, lexeme)
        elif lexeme == 'break':
            return Token(Class.BREAK, lexeme)
        elif lexeme == 'continue':
            return Token(Class.CONTINUE, lexeme)
        elif lexeme == 'return':
            return Token(Class.RETURN, lexeme)
        elif lexeme in ['integer', 'boolean', 'real', 'char', 'string', 'array']:
            return Token(Class.TYPE, lexeme)
        elif lexeme == 'begin':
            return Token(Class.BEGIN, lexeme)
        elif lexeme == 'end':
            return Token(Class.END, lexeme)
        elif lexeme == 'exit':
            return Token(Class.EXIT, lexeme)
        elif lexeme == 'var':
            return Token(Class.VAR, lexeme)
        elif lexeme == 'procedure':
            return Token(Class.PROCEDURE, lexeme)
        elif lexeme == 'function':
            return Token(Class.FUNCTION, lexeme)
        elif lexeme == 'downto':
            return Token(Class.DOWNTO, lexeme)
        #elif lexeme == 'array':
        #    return Token(Class.ARRAY, lexeme)
        elif lexeme == 'div':
            return Token(Class.DIV, lexeme)
        elif lexeme == 'mod':
            return Token(Class.MOD, lexeme)
        elif lexeme == 'and':
            return Token(Class.AND, lexeme)
        elif lexeme == 'or':
            return Token(Class.OR, lexeme)
        elif lexeme == 'xor':
            return Token(Class.XOR, lexeme)
          
        return Token(Class.ID, lexeme)

#Metoda next_char pomera pokazivač na sledeći karakter.
    def next_char(self):
        self.pos += 1
        if self.pos >= self.len:
            return None
        return self.text[self.pos]

#Metoda next_token konstruiše token odgovarajuće klase pozivajući metodu next_char.
    def next_token(self):
        self.read_space()
        curr = self.next_char()
        if curr is None:
            return Token(Class.EOF, curr)
        token = None
        if curr.isalpha():
            token = self.read_keyword()
        elif curr.isdigit():
            digit = self.read_int()
            if '.' not in digit:
              token = Token(Class.INT, digit)
            else:
              token = Token(Class.REAL, digit)
        elif curr == '\'':
            curr = self.next_char()
            if self.next_char == '\'':
              self.pos -= 1
              token = Token(Class.CHAR, self.read_char())
            else:
              self.pos -= 1
              token = Token(Class.STRING, self.read_string())
        elif curr == '+':
            token = Token(Class.PLUS, curr)
        elif curr == '-':
            token = Token(Class.MINUS, curr)
        elif curr == '*':
            token = Token(Class.STAR, curr)
        elif curr == '/':
            token = Token(Class.FWDSLASH, curr)
        elif curr == '&':
            curr = self.next_char()
            if curr == '&':
                token = Token(Class.AND, '&&')
            else:
                token = Token(Class.ADDRESS, '&')
                self.pos -= 1
        elif curr == '|':
            curr = self.next_char()
            if curr == '|':
                token = Token(Class.OR, '||')
            else:
                self.die(curr)
        elif curr == '=':
            token = Token(Class.EQ, '=')
        elif curr == ':':
            curr = self.next_char()
            if curr == '=':
              token = Token(Class.ASSIGN, ':=')  
            else:
              token = Token(Class.COLON, ':')
              self.pos -=1
        elif curr == '<':
            curr = self.next_char()
            if curr == '>':
                token = Token(Class.NEQ, '<>')  
            if curr == '=':
                token = Token(Class.LTE, '<=')
            else:
                token = Token(Class.LT, '<')
                self.pos -= 1
        elif curr == '>':
            curr = self.next_char()
            if curr == '=':
                token = Token(Class.GTE, '>=')
            else:
                token = Token(Class.GT, '>')
                self.pos -= 1
        elif curr == '(':
            token = Token(Class.LPAREN, curr)
        elif curr == ')':
            token = Token(Class.RPAREN, curr)
        elif curr == '[':
            token = Token(Class.LBRACKET, curr)
        elif curr == ']':
            token = Token(Class.RBRACKET, curr)
        elif curr == ';':
            token = Token(Class.SEMICOLON, curr)
        elif curr == ',':
            token = Token(Class.COMMA, curr)
        elif curr == '.':
            curr = self.next_char()
            if curr == '.':
              token = Token(Class.DOTDOT, '..')
            else:
              token = Token(Class.DOT, '.')
              self.pos -= 1
        else:
            self.die(curr)
        return token
        
#Metoda lex formira niz tokena pozivajući metodu next_token.
    def lex(self):
        tokens = []
        while True:
            curr = self.next_token()
            tokens.append(curr)
            if curr.class_ == Class.EOF:
                break
        return tokens

#Metoda die se koristi u slučaju da je lekser pročitao neočekivani karakter.
    def die(self, char):
        raise SystemExit("Unexpected character: {}".format(char))
```

# Parser

```python
class Node():
    pass


class Program(Node):
    def __init__(self, nodes):
        self.nodes = nodes


class Begin(Node):
    def __init__(self, block):
        self.block = block

class Decl(Node):
    def __init__(self, id_, type_):
        self.id_ = id_
        self.type_ = type_


class ArrayDecl(Node):
    def __init__(self, type_, id_, size, elems):
        self.type_ = type_
        self.id_ = id_
        self.size = size
        self.elems = elems


class ArrayElem(Node):
    def __init__(self, id_, index):
        self.id_ = id_
        self.index = index


class Assign(Node):
    def __init__(self, id_, expr):
        self.id_ = id_
        self.expr = expr


class If(Node):
    def __init__(self, cond, true, false):
        self.cond = cond
        self.true = true
        self.false = false


class Then(Node):
    def __init__(self, block):
        self.block = block


class While(Node):
    def __init__(self, cond, block):
        self.cond = cond
        self.block = block


class Repeat(Node):   #isto kao while
    def __init__(self, cond, block):
        self.cond = cond
        self.block = block


class For(Node):
    def __init__(self, to, block):
        #self.init1 = init
        self.to = to
        #self.init2 = init
        #self.do = do
        self.block = block


class FuncImpl(Node):
    def __init__(self, type_, id_, params, block):
        self.type_ = type_
        self.id_ = id_
        self.params = params
        self.block = block


class FuncCall(Node):
    def __init__(self, id_, args):
        self.id_ = id_
        self.args = args


class ProcImpl(Node):
    def __init__(self, id_, params, vars, block):
        self.id_ = id_
        self.params = params
        self.vars = vars
        self.block = block


class ProcCall(Node):
    def __init__(self, id_, args):
        self.id_ = id_
        self.args = args


class For(Node):
  def __init__(self, init1, init2, way, block):
    self.init1 = init1
    self.init2 = init2
    self.way = way
    self.block = block
        

class Block(Node):
    def __init__(self, nodes):
        self.nodes = nodes


class Params(Node):
    def __init__(self, params):
        self.params = params


class Args(Node):
    def __init__(self, args):
        self.args = args


class Elems(Node):
    def __init__(self, elems):
        self.elems = elems


class Break(Node):
    pass


class Continue(Node):
    pass


class Exit(Node):
    pass


class Type(Node):
    def __init__(self, value):
        self.value = value


class Int(Node):
    def __init__(self, value):
        self.value = value


class Real(Node):
    def __init__(self, value):
        self.value = value


class Char(Node):
    def __init__(self, value):
        self.value = value


class Boolean(Node):
    def __init__(self, value):
        self.value = value


class String(Node):
    def __init__(self, value):
        self.value = value


class Id(Node):
    def __init__(self, value):
        self.value = value


class Var(Node):
    def __init__(self, block):
        self.block = block


class BinOp(Node):
    def __init__(self, symbol, first, second):
        self.symbol = symbol
        self.first = first
        self.second = second


class UnOp(Node):
    def __init__(self, symbol, first):
        self.symbol = symbol
        self.first = first
```

```python
from functools import wraps
import pickle
```

```python
class Parser:
    def __init__(self, tokens):
        self.tokens = tokens
        self.curr = tokens.pop(0)
        self.prev = None

    def restorable(call):
        @wraps(call)
        def wrapper(self, *args, **kwargs):
            state = pickle.dumps(self.__dict__)
            result = call(self, *args, **kwargs)
            self.__dict__ = pickle.loads(state)
            return result
        return wrapper

    def eat(self, class_):
        if self.curr.class_ == class_:
            self.prev = self.curr
            self.curr = self.tokens.pop(0)
        else:
            self.die_type(class_.name, self.curr.class_.name)
  
    def program(self):
        nodes = []
        while self.curr.class_ != Class.EOF:
            if self.curr.class_ == Class.BEGIN:
                nodes.append(self.begin())
                if self.curr.class_ == Class.DOT:
                  self.eat(Class.DOT)
                else:
                  self.eat(Class.SEMICOLON)
            elif self.curr.class_ == Class.VAR:
                nodes.append(self.varDecl()) 
            elif self.curr.class_ == Class.PROCEDURE:
                nodes.append(self.procDecl())
            elif self.curr.class_ == Class.FUNCTION:
                nodes.append(self.funcDecl())
            else:
                self.die_deriv(self.program.__name__)
        return Program(nodes) 
   
    def begin(self):
        self.eat(Class.BEGIN)
        block = self.block()
        self.eat(Class.END)
        return Begin(block)
    
    def varDecl(self):
        self.eat(Class.VAR)
        block = self.var_()
        return Var(block)

    def procDecl(self):
        self.eat(Class.PROCEDURE)
        return self.proc_()
        
    def funcDecl(self):
        self.eat(Class.PROCEDURE)     
        return self.func_()

    def var_(self):
        nodes = []
        varNodes = []
        array = Elems([])
        while self.curr.class_ not in [Class.BEGIN, Class.FUNCTION, Class.PROCEDURE]:
            while self.curr.class_ != Class.COLON:
              id_ = self.id_()
              nodes.append(id_)
              if self.curr.class_ == Class.COMMA:
                  self.eat(Class.COMMA)
            self.eat(Class.COLON)
            
            type_ = self.type_()
            if type_.value == 'string':
              self.eat(Class.LBRACKET)
              size = self.curr.lexeme
              self.eat(Class.INT)
              self.eat(Class.RBRACKET)
              
              for id_ in nodes:
                varNodes.append(ArrayDecl(type_, id_, Int(size), Elems([])))

            elif type_.value == 'array':
              self.eat(Class.LBRACKET)
              x = int(self.curr.lexeme)
              self.eat(Class.INT)
              self.eat(Class.DOTDOT)
              y = int(self.curr.lexeme)
              #if (x is int) and (y is int):
              size = y - x + 1
              self.eat(Class.INT)
              self.eat(Class.RBRACKET)
              self.eat(Class.OF)
              type_ = self.type_()
              elems = None
              if self.curr.class_ == Class.EQ:
                self.eat(Class.EQ)
                self.eat(Class.LPAREN)
                elems = self.elems()
                self.eat(Class.RPAREN)
              for id in nodes:
                varNodes.append(ArrayDecl(type_, id, size, elems))
            else:
              for id in nodes:
                varNodes.append(Decl(id, type_))
            nodes = []
            self.eat(Class.SEMICOLON)
        return Block(varNodes)
            
    def id_(self):
        is_array_elem = self.prev.class_ != Class.TYPE
        id_ = Id(self.curr.lexeme)
        self.eat(Class.ID)
        #while self.curr.class_ == Class.COMMA:
        #    self.eat(Class.COMMA)
        #    self.eat(Class.ID)
        #    return id_
        if self.curr.class_ == Class.LPAREN:# and self.is_func_call(): 
            self.eat(Class.LPAREN)
            args = self.args()
            self.eat(Class.RPAREN)
            return FuncCall(id_, args)
        elif self.curr.class_ == Class.LBRACKET and is_array_elem:
            self.eat(Class.LBRACKET)
            index = self.expr()
            self.eat(Class.RBRACKET)
            id_ = ArrayElem(id_, index)    
        if self.curr.class_ == Class.ASSIGN:
            self.eat(Class.ASSIGN)
            expr = self.expr()
            if self.curr.class_ == Class.GTE or self.curr.class_ == Class.LTE or self.curr.class_ == Class.EQ:
              symbol = self.curr.lexeme
              self.eat(self.curr.class_)
              second = self.expr()
              return Assign(id_, BinOp(symbol, expr, second))
            else:
              return Assign(id_, expr)
        else:
            return id_
    
    def block(self):
        nodes = []
        while self.curr.class_ != Class.END:
            if self.curr.class_ == Class.IF:
                nodes.append(self.if_())
            elif self.curr.class_ == Class.WHILE:
                nodes.append(self.while_())
            elif self.curr.class_ == Class.FOR:
                nodes.append(self.for_())
            elif self.curr.class_ == Class.BREAK:
                nodes.append(self.break_())
            elif self.curr.class_ == Class.CONTINUE:
                nodes.append(self.continue_())
            elif self.curr.class_ == Class.RETURN:
                nodes.append(self.return_())
            elif self.curr.class_ == Class.VAR:
                nodes.append(self.var_())
            elif self.curr.class_ == Class.EXIT:
                nodes.append(self.exit_())
            elif self.curr.class_ == Class.ID:
                nodes.append(self.id_())
                #self.eat(Class.SEMICOLON)
            elif self.curr.class_ == Class.SEMICOLON:
                self.eat(Class.SEMICOLON)
            else:
                self.die_deriv(self.block.__name__)
        return Block(nodes)

    def proc_(self):
        id_ = self.id_()
        #self.eat(Class.ID)
        self.eat(Class.LPAREN)
        params = self.params()
        self.eat(Class.RPAREN)
        self.eat(Class.SEMICOLON)
        vars = None
        if self.curr.class_ == Class.VAR:
          vars = self.varDecl()
        self.eat(Class.BEGIN)
        block = self.block()
        self.eat(Class.END)
        self.eat(Class.SEMICOLON)
        return ProcImpl(id_, params, vars, block)
        
    def func_(self):
        id_ = self.id_()
        self.eat(Class.ID)
        self.eat(Class.LPAREN)
        params = self.params()
        self.eat(Class.RPAREN)
        if self.curr.class_ == Class.COLON:
          self.eat(Class.COLON)
        type_ = self.type_()
        self.eat(Class.SEMICOLON)
        #vars = []
        #if self.curr.class_ == Class.VAR:
        #  vars = self.varDecl()
        self.eat(Class.BEGIN)
        block = self.block()
        self.eat(Class.END)
        self.eat(Class.SEMICOLON)
        return FuncImpl(type_, id_, params, block)

    def if_(self):
        self.eat(Class.IF)
        cond = self.logic()
        self.eat(Class.THEN)
        self.eat(Class.BEGIN)
        true = self.block()
        self.eat(Class.END)
        false = None
        if self.curr.class_ == Class.ELSE:
            self.eat(Class.ELSE)
            self.eat(Class.BEGIN)
            false = self.block()
            self.eat(Class.END)
            self.eat(Class.SEMICOLON)
        #else:
        #   self.eat(Class.SEMICOLON)
        return If(cond, true, false)

    def while_(self):
        self.eat(Class.WHILE)
        self.eat(Class.LPAREN)
        cond = self.logic()
        self.eat(Class.RPAREN)
        self.eat(Class.DO)
        self.eat(Class.BEGIN)
        block = self.block()
        self.eat(Class.END)
        self.eat(Class.SEMICOLON)
        return While(cond, block)

    def repeat_(self):
        self.eat(Class.REPEAT)
        block = self.block()
        self.eat(Class.UNTIL)
        self.eat(Class.LPAREN)
        cond = self.logic()
        self.eat(Class.RPAREN)
        self.eat(Class.SEMICOLON)
        return Repeat(cond, block)
        
    def for_(self):
        self.eat(Class.FOR)

        init1 = self.id_()
        if self.curr.class_ == Class.TO:
          way = self.curr.lexeme
          self.eat(Class.TO)
        else:
          way = self.curr.lexeme
          self.eat(Class.DOWNTO)
        init2 = self.logic()
        self.eat(Class.DO)
        
        self.eat(Class.BEGIN)
        block = self.block()
        self.eat(Class.END)
        self.eat(Class.SEMICOLON)
        return For(init1, init2, way, block)

    def params(self):
        params = []
        nodes = []
        while self.curr.class_ != Class.COLON:
            if len(nodes) > 0:
              self.eat(Class.COMMA)
            id_ = self.id_()
            nodes.append(id_)
        self.eat(Class.COLON)
        type_ = self.type_()
        for id in nodes:
            params.append(Decl(type_, id))
        if self.curr.class_ == Class.SEMICOLON:
                self.eat(Class.SEMICOLON)
        return Params(params)

    def args(self):
        args = []
        while self.curr.class_ != Class.RPAREN:
            
            if len(args) > 0:
                self.eat(Class.COMMA)
            args.append(self.expr())

            if self.curr.class_ == Class.COLON:
                args.append(':')
                self.eat(Class.COLON)
                args.append(self.expr())
                args.append(':')
                self.eat(Class.COLON)
                args.append(self.expr())
            
        return Args(args)

    def elems(self):
        elems = []
        while self.curr.class_ != Class.RBRACE:
            if len(elems) > 0:
                self.eat(Class.COMMA)
            elems.append(self.expr())
        return Elems(elems)

    def exit_(self):
        self.eat(Class.EXIT)
        self.eat(Class.SEMICOLON)
        return Exit()

    def break_(self):
        self.eat(Class.BREAK)
        self.eat(Class.SEMICOLON)
        return Break()

    def continue_(self):
        self.eat(Class.CONTINUE)
        self.eat(Class.SEMICOLON)
        return Continue()

    def type_(self):
        type_ = Type(self.curr.lexeme)
        self.eat(Class.TYPE)
        return type_

    def factor(self):
        if self.curr.class_ == Class.INT:
            value = Int(self.curr.lexeme)
            self.eat(Class.INT)
            return value
        elif self.curr.class_ == Class.REAl:
            value = Real(self.curr.lexeme)
            self.eat(Class.REAL)
            return value
        elif self.curr.class_ == Class.CHAR:
            value = Char(self.curr.lexeme)
            self.eat(Class.CHAR)
            return value
        elif self.curr.class_ == Class.BOOLEAN:
            value = Boolean(self.curr.lexeme)
            self.eat(Class.BOOLEAN)
            return value
        elif self.curr.class_ == Class.STRING:
            value = String(self.curr.lexeme)
            self.eat(Class.STRING)
            return value
        elif self.curr.class_ == Class.ID:
            return self.id_()
        elif self.curr.class_ in [Class.MINUS, Class.NOT]:
            op = self.curr
            self.eat(self.curr.class_)
            first = None
            if self.curr.class_ == Class.LPAREN:
                self.eat(Class.LPAREN)
                first = self.logic()
                self.eat(Class.RPAREN)
            else:
                first = self.factor()
            return UnOp(op.lexeme, first)
        elif self.curr.class_ == Class.LPAREN:
            self.eat(Class.LPAREN)
            first = self.logic()
            self.eat(Class.RPAREN)
            return first
        elif self.curr.class_ == Class.SEMICOLON:
            return None
        else:
            self.die_deriv(self.factor.__name__)

    def term(self):
        first = self.factor()
        while self.curr.class_ in [Class.STAR, Class.FWDSLASH, Class.DIV, Class.MOD]: # Class.TO, Class.DO]:
            if self.curr.class_ == Class.STAR:
                op = self.curr.lexeme
                self.eat(Class.STAR)
                second = self.factor()
                first = BinOp(op, first, second)
            elif self.curr.class_ == Class.FWDSLASH:
                op = self.curr.lexeme
                self.eat(Class.FWDSLASH)
                second = self.factor()
                first = BinOp(op, first, second)
            elif self.curr.class_ == Class.DIV:
                op = self.curr.lexeme
                self.eat(Class.DIV)
                second = self.factor()
                first = BinOp(op, first, second)
            elif self.curr.class_ == Class.MOD:
                op = self.curr.lexeme
                self.eat(Class.MOD)
                second = self.factor()
                first = BinOp(op, first, second)
        return first

    def expr(self):
        first = self.term()
        while self.curr.class_ in [Class.PLUS, Class.MINUS]:
            if self.curr.class_ == Class.PLUS:
                op = self.curr.lexeme
                self.eat(Class.PLUS)
                second = self.term()
                first = BinOp(op, first, second)
            elif self.curr.class_ == Class.MINUS:
                op = self.curr.lexeme
                self.eat(Class.MINUS)
                second = self.term()
                first = BinOp(op, first, second)
        return first

    def compare(self):
        first = self.expr()
        if self.curr.class_ == Class.EQ:
            op = self.curr.lexeme
            self.eat(Class.EQ)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.NEQ:
            op = self.curr.lexeme
            self.eat(Class.NEQ)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.LT:
            op = self.curr.lexeme
            self.eat(Class.LT)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.GT:
            op = self.curr.lexeme
            self.eat(Class.GT)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.LTE:
            op = self.curr.lexeme
            self.eat(Class.LTE)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.GTE:
            op = self.curr.lexeme
            self.eat(Class.GTE)
            second = self.expr()
            return BinOp(op, first, second)
        else:
            return first

    def logic(self):
        first = self.compare()
        if self.curr.class_ == Class.OR:
            op = self.curr.lexeme
            self.eat(Class.OR)
            second = self.compare()
            first = BinOp(op, first, second)
        elif self.curr.class_ == Class.AND:
            op = self.curr.lexeme
            self.eat(Class.AND)
            second = self.compare()
            first = BinOp(op, first, second)
        elif self.curr.class_ == Class.XOR:
            op = self.curr.lexeme
            self.eat(Class.XOR)
            second = self.compare()
            first = BinOp(op, first, second)
        else:
            return first

    #@restorable
    #def is_func_call(self):
       # try:
       #     boo = False
       #     self.eat(Class.LPAREN)
       #     self.args()
       #     self.eat(Class.RPAREN)

       #     if self.curr.class_ == Class.SEMICOLON:
       #       self.eat(Class.SEMICOLON)
       #       if self.curr.class_ == Class.BEGIN or self.curr.class_ == Class.VAR:
       #        boo = False
            
       #     if self.curr.class_ != Class.COLON:
       #       boo = True
            
       #     return boo
       # except:
       #     return False

    def parse(self):
        return self.program()

    def die(self, text):
        raise SystemExit(text)

    def die_deriv(self, fun):
        self.die("Derivation error: {}".format(fun))

    def die_type(self, expected, found):
        self.die("Expected: {}, Found: {}".format(expected, found))
```

# Generator

```python
import re
from graphviz import Digraph, Source
from IPython.display import Image
from functools import wraps
import pickle

class Visitor():
    def visit(self, parent, node):
        method = 'visit_' + type(node).__name__
        visitor = getattr(self, method, self.die)
        return visitor(parent, node)

    def die(self, parent, node):
        method = 'visit_' + type(node).__name__
        raise SystemExit("Missing method: {}".format(method))
```

```python
class Generator(Visitor):
    def __init__(self, ast):
        self.ast = ast
        self.c = ""
        self.level = 0
        self.id_type = list()

    def append(self, text):
        self.c += str(text) 

    def newline(self):
        self.append('\n\r')

    def indent(self):
        for i in range(self.level):
            self.append('\t')

    def visit_Program(self, parent, node):
        for n in node.nodes:
          if type(n).__name__ == 'Var':
            self.append('int main() {')
            self.newline()
          self.visit(node, n)
        self.newline()
        self.level += 1
        self.newline()
        self.indent()
        self.append('return 0;')
        self.newline()
        self.append('}')

    def visit_Var(self, parent, node):
        self.visit(node, node.block)
        self.newline()

    def visit_Begin(self, parent, node):#if !var append main()
        self.visit(node, node.block)
        self.newline()

    def visit_Decl(self, parent, node):
        idType = (node.id_, node.type_)
        self.id_type.append(idType)
        self.visit(node, node.type_)
        self.append(' ')
        #if(','):
        self.visit(node, node.id_)
        #self.append(', ')
        self.append(';')

    def visit_NoneType(self, parent, node):
        parent = node

    def visit_ArrayDecl(self, parent, node):
        idType = (node.id_, node.type_)
        self.id_type.append(idType)
        self.visit(node, node.type_)
        self.append(' ')
        self.visit(node, node.id_)
        self.append('[ ')
        self.append(node.size) # append?
        self.append(' ]')
        self.append(';')

    def visit_ArrayElem(self, parent, node):
        self.visit(node, node.id_)
        self.append('[')
        self.visit(node, node.index)
        self.append(']')

    def visit_Assign(self, parent, node):
        self.visit(node, node.id_)
        self.append(' = ')
        self.visit(node, node.expr)
        self.append(';')

    def visit_If(self, parent, node):
        self.append('if (')
        self.visit(node, node.cond)
        self.append(')')
        self.append('{')
        self.newline()
        self.visit(node, node.true)
        self.indent()
        self.append('}')
        if node.false is not None:
            self.append('else {')
            self.newline()
            self.visit(node, node.false)
            self.indent()
            self.append('}')

    def visit_While(self, parent, node):
        self.append('while ')
        self.visit(node, node.cond)
        self.append(':')
        self.newline()
        self.visit(node, node.block)

    def visit_For(self, parent, node):
        self.newline()
        self.indent()
        self.append('for (')
        self.visit(node, node.init1)
        if node.way == 'downto':
          self.visit(node, node.init1.id_)
          self.append(' >= ')
          self.visit(node, node.init2)
          self.append('; ')
          self.visit(node, node.init1.id_)
          self.append('-- ')
        else:
          self.visit(node, node.init1.id_)
          self.append(' <= ')
          self.visit(node, node.init2)
          self.append('; ')
          self.visit(node, node.init1.id_)
          self.append('++ ')
        self.append(') {')
        self.newline()
        self.visit(node, node.block)
        self.newline()
        self.indent()
        self.append('}')
        self.newline()
        #self.append('while ')
        #self.visit(node, node.cond)
        #self.append(':')
        #self.newline()
        #self.visit(node, node.block)
        #self.level += 1
        #self.indent()
        #self.visit(node, node.step)
        #self.level -= 1

    def visit_FuncImpl(self, parent, node):
        idType = (node.id_, node.type_)
        self.idTypeFuncTuples.append(idType)
        self.visit(node, node.type_)
        self.append(' ')
        self.append(node.id_.value)
        self.append('( ')
        self.visit(node, node.params)
        self.append(' )')
        self.newline()
        self.indent()
        self.visit(node, node.block)
        self.newline()

    def visit_FuncCall(self, parent, node):
        func = node.id_.value
        args = node.args.args
        
        if func == 'write' or func == 'writeln':
          
          vars = []
          #format_ = args[0].value
          #matches = re.findall('%[dcs]', format_)
          #format_ = re.sub('%[dcs]', '{}', format_)
          self.append('printf("')
          #self.append(format_)
          #self.append('"')

          if func == 'writeln':
              self.append('\\n')
              
          for a in args:
            
            if type(a).__name__ == 'String':
                self.append(a.value)

            elif type(a).__name__ == 'Char':
                self.append(a.value)

            elif type(a).__name__ == 'BinOp':
                vars.append(a)

                for id_, type_ in self.id_type:
                  if id_.value == a.first.value:
                    if type_.value == 'string':
                      self.append('%s')
                    elif type_.value == 'char':
                      self.append('%c')
                    elif type_.value == 'integer':
                      self.append('%d')
                    elif type_.value == 'real':
                      self.append('%.2f')
                    elif type_.value == 'boolean':
                      self.append('%d')
            
            elif type(a).__name__ == 'FuncCall':
              vars.append(a)

              if a.id_.value == 'chr':
                  self.append('%c')
              elif a.id_.value in ['inc','dec','ord']:
                  self.append('%d')
        

            elif type(a).__name__ == 'Id':
                vars.append(a)

                for id_, type_ in self.id_type:
                  if id_.value == a.value:
                    if type_.value == 'char':
                      self.append('%c')
                    elif type_.value == 'string':
                      self.append('%s')
                    elif type_.value == 'integer':
                      self.append('%d')
                    elif type_.value == 'real':
                      self.append('%f')
                    elif type_.value == 'boolean':
                      self.append('%d')

            elif type(a).__name__ == 'ArrayElem':
              vars.append(a)

              for id_, type_ in self.id_type:
                if id_.value == a.id_.value:
                  if type_.value == 'string':
                    self.append('%s')
                  elif type_.value == 'char':
                    self.append('%c')
                  elif type_.value == 'integer':
                    self.append('%d')
                  elif type_.value == 'real':
                    self.append('%f')
                  elif type_.value == 'boolean':
                    self.append('%d')
            
          self.append('"')
          if len(vars) > 0:
              self.append(', ')
          
          for i, var in enumerate(vars):
            self.visit(node, var)

            if i != (len(vars) - 1):
              self.append(', ')
          self.append(');')

        elif func == 'read' or func == 'readln':
          
          self.append('scanf("')
          
          for a in args:
            for id_, type_ in self.id_type:
              if type(a).__name__ != 'ArrayElem':
                if id_.value == a.value:
                  if type_.value == 'string':
                    self.append('%s')
                  elif type_.value == 'char':
                    self.append('%c')
                  elif type_.value == 'integer':
                    self.append('%d')
                  elif type_.value == 'real':
                    self.append('%f')
                  elif type_.value == 'boolean':
                    self.append('%d')
              else:
                if id_.value == a.id_.value:
                  if type_.value == 'string':
                    self.append('%s')
                  elif type_.value == 'char':
                    self.append('%c')
                  elif type_.value == 'integer':
                    self.append('%d')
                  elif type_.value == 'real':
                    self.append('%f')
                  elif type_.value == 'boolean':
                    self.append('%d')
                
            
          self.append('", ')

          for i, a in enumerate(args):
            
            self.append('&')
            self.visit(node, a)
            if i != (len(args) - 1):
              self.append(', ')
          self.append(');')
        
        elif func == 'chr':
            self.visit(node.args, args[0])
            
        elif func == 'ord':
            self.visit(node.args, args[0])

        else:
          self.append(func)
          self.append('(')
          self.visit(node, node.args)
          self.append(');')

    def visit_ProcImpl(self, parent, node):
        self.append('void')
        self.append(' ')
        self.append(node.id_.value)
        self.append('(')
        self.visit(node, node.params)
        self.append(')')
        self.append('{')
        self.newline()
        self.indent()
        self.visit(node, node.block)
        self.append('}')
        self.newline()

    def visit_Block(self, parent, node):
        self.level += 1
        for n in node.nodes:
            self.indent()
            self.visit(node, n)
            self.newline()
        self.level -= 1

    def visit_Params(self, parent, node):
        for i, p in enumerate(node.params):
            if i > 0:
                self.append(', ')
            self.visit(p, p.id_)
            self.append(' ')
            self.visit(p, p.type_)

    def visit_Args(self, parent, node):
        for i, a in enumerate(node.args):
            if i > 0:
                self.append(', ')
            self.visit(node, a)

    def visit_Elems(self, parent, node):
        for i, e in enumerate(node.elems):
            if i > 0:
                self.append(', ')
            self.visit(node, e)

    def visit_Break(self, parent, node):
        self.append('break')

    def visit_Continue(self, parent, node):
        self.append('continue;')

    def visit_Exit(self, parent, node):#ne treba
        self.append('return;')
        #if node.expr is not None:
        #  self.append(' ')
        #  self.visit(node, node.expr)

    def visit_Type(self, parent, node):
        if node.value == 'integer':
            self.append('int')
        elif node.value == 'string':
            self.append('char')
        elif node.value == 'real':
            self.append('float')
        elif node.value == 'boolean':
            self.append('int')
        else :
            self.append(node.value)

    def visit_Int(self, parent, node):
        self.append(node.value)

    def visit_Real(self, parent, node):
        self.append(node.value)

    def visit_Char(self, parent, node):
        self.append(ord(node.value)) 

    def visit_Boolean(self, paremt, node):
        if node.value == 'true':
          self.append('1')
        elif node.value == 'false':
          self.append('0')

    def visit_String(self, parent, node):
        self.append(node.value)

    def visit_Id(self, parent, node):
        self.append(node.value)

    def visit_BinOp(self, parent, node):
        self.visit(node, node.first)
        self.append(' ')
        if node.symbol == 'and':
            self.append(' && ')
        elif node.symbol == 'or':
            self.append('||')
        elif node.symbol == 'div':
            self.append('/')
        elif node.symbol == 'mod':
            self.append('%')
        elif node.symbol == '=':
            self.append('==')
        elif node.symbol == '<>':
            self.append('!=')
        else:
            self.append(node.symbol)
        self.append(' ')
        self.visit(node, node.second)

    def visit_UnOp(self, parent, node):
        if node.symbol == 'not':
            self.append('!')
        elif node.symbol != '&':
            self.append(node.symbol)
        self.visit(node, node.first)

    def generate(self, path):
        self.visit(None, self.ast)
        self.c = re.sub('\n\s*\n', '\n', self.c)
        with open(path, 'w') as source:
            source.write(self.c)
        return path
```

# Symbolizer


```python
class Symbol:
    def __init__(self, id_, type_, scope):
        self.id_ = id_
        self.type_ = type_
        self.scope = scope

    def __str__(self):
        return "<{} {} {}>".format(self.id_, self.type_, self.scope)

    def copy(self):
        return Symbol(self.id_, self.type_, self.scope)
```

```python
class Symbols:
    def __init__(self):
        self.symbols = {}

    def put(self, id_, type_, scope):
        self.symbols[id_] = Symbol(id_, type_, scope)

    def get(self, id_):
        return self.symbols[id_]

    def contains(self, id_):
        return id_ in self.symbols

    def remove(self, id_):
        del self.symbols[id_]
    
    def __len__(self):
        return len(self.symbols)

    def __str__(self):
        out = ""
        for _, value in self.symbols.items():
            if len(out) > 0:
                out += "\n"
            out += str(value)
        return out

    def __iter__(self):
        return iter(self.symbols.values())

    def __next__(self):
        return next(self.symbols.values())
```

```python
class Symbolizer(Visitor):
    def __init__(self, ast):
        self.ast = ast

    def visit_Program(self, parent, node):
        node.symbols = Symbols()
        for n in node.nodes:
            self.visit(node, n)

    def visit_Decl(self, parent, node):
        parent.symbols.put(node.id_.value, node.type_.value, id(parent))

    def visit_Var(self, parent, node):
        node.symbols = Symbols()
        self.visit(node, node.block)
        
    def visit_Begin(self, parent, node):
        node.symbols = Symbols()
        self.visit(node, node.block)
            
    def visit_ArrayDecl(self, parent, node):
        node.symbols = Symbols()
        parent.symbols.put(node.id_.value, node.type_.value, id(parent))

    def visit_ArrayElem(self, parent, node):
        pass

    def visit_Assign(self, parent, node):
        pass
    
    def visit_If(self, parent, node):
        self.visit(node, node.true)
        if node.false is not None:
            self.visit(node, node.false)

    def visit_While(self, parent, node):
        self.visit(node, node.block)

    def visit_For(self, parent, node):
        self.visit(node, node.block)

    def visit_FuncImpl(self, parent, node):
        parent.symbols.put(node.id_.value, node.type_.value, id(parent))
        self.visit(node, node.block)
        self.visit(node, node.params)

    def visit_FuncCall(self, parent, node):
        pass

    def visit_ProcImpl(self, parent, node):
        parent.symbols.put(node.id_.value, 'void', id(parent))
        self.visit(node, node.block)
        self.visit(node, node.params)

    def visit_Block(self, parent, node):
        node.symbols = Symbols()
        for n in node.nodes:
            self.visit(node, n)

    def visit_Params(self, parent, node):
        node.symbols = Symbols()
        for p in node.params:
            self.visit(node, p)
            self.visit(parent.block, p)

    def visit_Args(self, parent, node):
        pass

    def visit_Elems(self, parent, node):
        pass

    def visit_Break(self, parent, node):
        pass

    def visit_Continue(self, parent, node):
        pass

    def visit_Return(self, parent, node):
        pass

    def visit_Type(self, parent, node):
        pass

    def visit_Int(self, parent, node):
        pass

    def visit_Char(self, parent, node):
        pass

    def visit_String(self, parent, node):
        pass

    def visit_Id(self, parent, node):
        pass

    def visit_Exit(self, parent, node):
        pass
        
    def visit_BinOp(self, parent, node):
        pass

    def visit_UnOp(self, parent, node):
        pass

    def symbolize(self):
        self.visit(None, self.ast)
```
